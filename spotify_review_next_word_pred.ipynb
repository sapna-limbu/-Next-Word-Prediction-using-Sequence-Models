```python
from tensorflow.keras.preprocessing.text import Tokenizer

tokenizer =Tokenizer()
```

```python
tokenizer.fit_on_texts(df['content'][0:1000])
```

```python
import pickle 
```

```python
# Save the tokenizer to a file
with open('tokenizer.pkl', 'wb') as handle:
    pickle.dump(tokenizer, handle, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
len(tokenizer.word_index)
```

```python
data = tokenizer.texts_to_sequences(df['content'][0:1000])
```

```python
inputs=[]
output=[]
for token in data:
    for index in range(len(token)-1):
        inputs.append(token[:index+1])
        output.append(token[index+1])
```

```python
print(len(inputs),len(output))
```

```python
from keras.utils import pad_sequences, to_categorical
```

```python
feature_variable= pad_sequences(inputs, padding="pre")
```

```python
feature_variable.shape
```

```python
class_label= to_categorical(output)
```

```python
class_label.shape
```

```python
len(output)
```

```python
model=Sequential()
model.add(InputLayer(shape=(105,)))
model.add(Embedding(2258,5))
model.add(Bidirectional(LSTM(40,return_sequences=False)))
model.add(Dense(2258,activation="softmax"))
```

```python
model.compile(optimizer='rmsprop',
    loss="categorical_crossentropy",
    metrics=["accuracy"])
```

```python
model.fit(feature_variable,class_label,batch_size= 120, epochs=2000)
```

```python
import  time
x=input("Enter a seed word: ")
for y in range(20):
   word=tokenizer.index_word[np.argmax(model.predict(pad_sequences(tokenizer.texts_to_sequences([x]),maxlen=106)))]
   print(x)
   x=x+" "+word
   time.sleep(0.9)
```
